### 1.1 Helper functions for data collections

```python
import json

def save_json_file(file_name, obj):  #"speech", "speaker" #text_lst ,strong_text_lst
    json_data = json.dumps(obj) #text_lst #strong_text_lst
    # Write to a file
    with open(f'{file_name}.json', 'w') as json_file:
        json_file.write(json_data)
        
def load_json_file(json_file_name):#"speaker" "speech"
    with open(f'{json_file_name}.json', 'r') as file:
        data = json.load(file)
    return data
```

### 2.1 Helper functions

```python
from collections import defaultdict

def merge_similar_keys(data): #Group the speeches by speakers
    merged_dict = defaultdict(list)
    for key, value in data.items():
        normalized_key = key.split(' (')[0]
        merged_dict[normalized_key].extend(value)
    return dict(merged_dict)
```

About: Minimal keyword extraction technique that leverages BERT embeddings to find key phrases and words in a document. 
- Based on the principle that the contextual embeddings generated by models like BERT can capture semantic meanings of words and phrases in a way that traditional keyword extraction methods, which often rely on statistical measures, cannot

1. Input document embedded using a pre-trained BERT model. You can pick any BERT model your want from 🤗transformers (sidenote: may want to experiment on different versions of BERT; RoBERTa, ELECTRA, AlBERT, DeBERTa, DistilBERT etc. - Turns chunk of text into fixed-sized vector representing semantic aspect of the document

2. Keywords and expressions (n-grams) are extracted from the same document using Bag Of Words techniques (such as a TfidfVectorizer or CountVectorizer).

3. Each keyword is then embedded into a fixed-size vector with the same model used to embed the document

4. Now that the keywords and the document are represented in the same space, KeyBERT computes a cosine similarity between the keyword embeddings and the document embedding. Then, the most similar keywords (with the highest cosine similarity score) are extracted.

Includes 2 methods to introduce <b>diversity</b> 

1 — Max Sum Similarity (MSS)

- Setting the top_n argument to a value, say 20. 
- 2 x top_n keywords are extracted from the document. 
- Pairwise similarities are computed between these keywords.
- Method extracts the most relevant keywords that are the least similar to each other.




2 - Maximal Marginal Relevance (MMR) ---- explained further later

Limitation:
- Execution time - large documents and real-time results
- Use smaller models like DistilBERT, mixed precision, convert to ONNX format

2. KeyBERT then extracts simple word n-grams of the defined length from documents and uses them as candidate keyphrases for embedding creation and similarity calculation

How to enhance 
- KeyphraseVectorizers package extracts keyphrases with part-of-speech patterns from a collection of text documents and converts them into a document-keyphrase matrix.
- Matematical matrix describes frequency of keyphrases occuring in collection of documents

How KeyphaseVectorizer package work?
1. Document texts are annotated with spaCy part-of-speech tags
2. Keyphrases are extracted from the document texts whose part-of-speech tags match a predefined regex pattern
3. Vectorizers(Default, the vectorizers extract keyphrases with zero or more adjectives, followed by one or more nouns using the English spaCy part-of-speech tags) calculate document-keyphrase matrices                                                         

### KeyBERTInspired https://maartengr.github.io/BERTopic/api/representation/keybert.html

Follows KEYBERT but does some optimization (never specify)
1) Extract the top n representative documents per topic
2) Randomly sample a number of candidate documents per cluster which is controlled by the nr_samples parameter
3) Top n representative documents are extracted by calculating the c-TF-IDF representation for the candidate via <b>cosine similarity </b> closest to the topic c-TF-IDF representation
4) The top n words per topic are extracted based on their c-TF-IDF representation, which is controlled by the nr_repr_docs parameter 
5) Extract the embeddings for words and representative documents and create topic embeddings by averaging the representative documents
6) Most similar words to each topic are extracted by calculating the cosine similarity between word and topic embeddings

### MaximalMarginalRelevance (MMR)

About - MMR considers the similarity of keywords/keyphrases with the document, along with the similarity of already selected keywords and keyphrases. This results in a selection of keywords that maximize their within diversity with respect to the document



https://medium.com/tech-that-works/maximal-marginal-relevance-to-rerank-results-in-unsupervised-keyphrase-extraction-22d95015c7c5
- Tries to reduce redundancy and increase diversity in the result and is used in text summarization
- MMR selects the phrase in the final keyphrases list according to a combined criterion of query relevance and novelty of information
- Considers the similarity of keyphrase with the document, along with the similarity of already selected phrases

<b>Advantage:</b>
1. Balance Between Relevance and Diversity: MMR effectively balances relevance with diversity, ensuring that the selected items are not just relevant but also varied. This is particularly useful in search engines and summarization tasks where presenting diverse perspectives or aspects is crucial.

<b>Disadvatnage:</b>
1. Computational cost
2. Potential Loss of Highly Relevant Items: In its effort to maintain diversity, MMR might exclude some highly relevant items in favor of less relevant but more diverse ones. This trade-off might not always be desirable, depending on the context.
3. Subjectivity in Diversity: What constitutes "diversity" can be subjective and context-dependent. Defining and quantifying diversity appropriately for a given application can be challenging.
